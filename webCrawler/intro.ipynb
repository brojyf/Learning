### Requests

```python
import requests

URL = "https://www.plu.edu/dining/menus/lunch/"
response = requests.get(URL)

response
```

### BeautifulSoup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "lxml")
divs = soup.find_all("div", class_="plu-food-menu-item")
divs[0]
```

```python
# Get each restaurant
crave = []
cross_cultures = {}
aglio = {}
for div in divs:
    crave_temp = div.find_all(
        'div',
        attrs={'data-restaurant-name': 'crave'}
    )
    crave.append(crave_temp)

crave
```

```python
for div in divs:
    names = div.find_all('h4', class_='item-name')
    for name in names:
        print(name.get_text(strip=True))
```
